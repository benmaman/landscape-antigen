# 1. Allign the Hertz flue antigen
in the next section we use the sequences of the Hertz antigen and the article antigen and try to allign them and find identitcal - close to identitcal

```python

df_article_antigen=pd.read_excel("list_of_antigen_hertz.xlsx",sheet_name='article_antigen')
df_hertz_antigen=pd.read_excel("list_of_antigen_hertz.xlsx",sheet_name='rha_antigen_Hertz')

#quick preprocess
df_hertz_antigen=df_hertz_antigen.loc[:,['name','seq']]

#replace '-' to null
df_article_antigen=df_article_antigen.replace("-",np.nan)
df_hertz_antigen=df_hertz_antigen.replace("-",np.nan)

df_article_antigen=df_article_antigen.replace(" ","")

df_hertz_antigen["seq"] = df_hertz_antigen["seq"].str.replace(r"\s+", "", regex=True).str.upper()
df_article_antigen["seq"] = df_article_antigen["seq"].str.replace(r"\s+", "", regex=True).str.upper()


df_hertz_antigen.dropna(inplace=True)
df_article_antigen.dropna(inplace=True)

```

```python
# Setup scoring
blosum62 = substitution_matrices.load("BLOSUM62")
gap_open = -10
gap_extend = -0.5
identity_thresh=0.96
```

```python
# Initialize result columns
df_hertz_antigen["similiar_seq_name"] = np.nan
df_hertz_antigen["similiar_seq"] = np.nan
df_hertz_antigen["score"] = np.nan
df_hertz_antigen["percent_identity"] = np.nan

# Loop through query sequences
for i, query in df_hertz_antigen["seq"].items():
    best_score = -np.inf
    best_seq = None
    best_name = None
    best_identity = 0

    for j, candidate in df_article_antigen["seq"].items():
        alignment = pairwise2.align.globalds(query, candidate, blosum62, gap_open, gap_extend)[0]
        score = alignment.score

        # Percent identity calculation
        aligned_q = alignment.seqA
        aligned_c = alignment.seqB
        matches = sum(1 for a, b in zip(aligned_q, aligned_c) if a == b and a != '-' and b != '-')
        aligned_len = sum(1 for a, b in zip(aligned_q, aligned_c) if a != '-' and b != '-')
        identity = matches / aligned_len if aligned_len > 0 else 0

        if score > best_score:
            best_score = score
            best_seq = candidate
            best_name = df_article_antigen.loc[j, "name"]
            best_identity = identity

    df_hertz_antigen.at[i, "similiar_seq"] = best_seq
    df_hertz_antigen.at[i, "similiar_seq_name"] = best_name
    df_hertz_antigen.at[i, "score"] = best_score
    df_hertz_antigen.at[i, "percent_identity"] = best_identity
```

```python
df_hertz_antigen[df_hertz_antigen["percent_identity"]>identity_thresh].to_csv("list_of_alligned_seqs.csv")
```
